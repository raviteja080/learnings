## Creating a Pie chart with the given data.

```python
# importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Cause of deaths 
cause = ['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis','Others']
```

```python
# percentage
percentage=[62, 5, 4, 2, 1 , 26 ]
```

```python
# creating a pie plot
plt.figure(figsize=(8,8))
explode=(0.03,0.01,0.01,0.01,0.01,0.01)
plt.pie(percentage,labels=cause,explode=explode,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title('Major causes of death in Ohio state in 2012')
plt.show
```
