### Lets start with introduction to Matplotlib

Matplotlib is again divided into three classes:
Backend Layer is divided into :
1. Figure Canvas: matplotlib.backend_bases.FigureCanvas
2. Rendere       : matplotlib.backend_bases.Renderer
3. Event :  matplotlib.backend_cases.Event

Artist Layer is again divided into 2 main object: 
1. Primitive : Line 2D, Rectangle, Circle and Text
2. Composite  : Axis, tick, Axes and Figure

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas #import figure Canvas
```

```python
from matplotlib.figure import Figure # import Figure Artist 
fig=Figure()
canvas=FigureCanvas(fig)

#create 10000 random numbers using numpy 
import numpy as np 
x=np.random.randn(10000)
ax=fig.add_subplot(111)
ax.hist(x,100)# 100 means create 100 bins 
ax.set_title("Normal Distribution with $\mu=0, \sigma=1$")
fig.savefig("matplotlib_histogram.png")
```

```python
# now using the scripting layer to the same task we did above we get 
import matplotlib.pyplot as plt
plt.hist(x,100)
plt.title("Normal Distribution with $\mu=0, \sigma=1$")
plt.savefig("matplotlib_histogram1.png")
plt.show()
```

# Visualizing Data using Matplotlib

```python
# importing libraries
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline 
```

```python
mpl.__version__
```

```python
df.loc['Haiti', years].plot() # lineplot for population arriving from Haiti
```

```python
haiti=df.loc['Haiti',years]
haiti.head()
```

```python
# changing the value of index of haiti to make it easy to plot
haiti.index=haiti.index.map(int)
haiti.plot(kind='line')
plt.title('Immigrants from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
# create a new data frame for India and China 
df1=df.loc[['India','China'],years]
df1.head()
```

```python
df1.plot(kind='line')  #defining the plot type as line
```

That is strange
Pandas plots indices on X axis and columns on Y axis
So we need to make Transpose of the df1 dataframe and then plot it to get a better plot

```python
df1=df1.T
df1.head
```

```python
df1.plot(kind='line')
plt.title(" Immigrants from India and China")
plt.xlabel("Years")
plt.ylabel("Number of immigrants")
```

```python
# lets check the data for the top 5 contributing to the immigration to Canada
df.sort_values(by='Total',ascending=False,axis=0,inplace=True)
```

```python
df.head(5)
```

```python
df5=df.head(5) #takes the top 5 countries into account
df5=df5[years].T
df5
```

```python
df5.index=df5.index.map(int)
df5.plot(kind='line',figsize=(14,8))

plt.title("Immigration trend of top 5 countries")
plt.xlabel("Years")
plt.ylabel("Number of immigrants")
plt.show()
```
